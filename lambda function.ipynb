```python
# Write a Python program that sums the length of a list of names after removing those that start with lowercase letters. Use the lambda function.


name=['Ravi','Ankit','Suman','deepak','manish']


fitered_name=list(filter(lambda x:x[0].isupper(),name))

total_length=sum(map(lambda x:len(x),fitered_name))

print("name sfter removing lowercase onces:",fitered_name)
print("sum of their length ",total_length)
```
