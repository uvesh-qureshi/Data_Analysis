```python
# Encapsulation in Python:

# Encapsulation is an Object-Oriented Programming (OOP) concept where data (variables) 
# and methods(functions)are wrapped inside a class and direct access to data is restricted.
# It helps in data hiding and security.

# In Python, encapsulation is achieved using access modifiers
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name        # public
        self.__marks = marks    # private variable

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.__marks)

s2 = Student("Rahul", 85)
s2.display()
```

```python
# It cannot be accessed directly outside the class.
print(s2.__marks)
```

```python
class Student21:
    def __init__(self):
        self.__marks = 0

    def setMarks(self, m):           #setMarks() → set value
        self.__marks = m

    def getMarks(self):             #getMarks() → get value                      
        return self.__marks

s = Student21()
s.setMarks(90)
print("Marks:", s.getMarks())
```

```python
class BankAccount_1:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def show_balance(self):
        print("Balance:", self.__balance)

a= BankAccount_1(5000)
a.show_balance()
```
