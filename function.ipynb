```python
# write a python function that takes a list and return a new list with distinct element fron the first list
def unique_list(first):
    new_list=[]
    
    
    for item in first:
        if item not in new_list:
            new_list.append(item)
    return new_list       


numbers=[1,2,3,3,4,4,5,6,7,8]
print(unique_list(numbers)) 




```
